```python
import numpy as np
print("\nFormula for Cross-Entropy Loss Function")
print("For positives")
print("1. -log(x)") 
print("For negatives")
print("2. -log(1-x)") 

print("\nFormula for Focal Loss (Alpha Form)")
print("alpha = 0.25")
print("gamma = 2")
print("1. -alpha * (1 - x)^gamma * log(x)") 
print("2. -(1-alpha) * x^gamma * log(1-x)")

print("\nCase 1 ::: Easy Example(Correctly Classified, y=1, x~1)")
print("Foreground with x = 0.9")
x = 0.9
alpha = 0.25
gamma = 2
c1 = -np.log(x)
f1 = -alpha * np.power((1 - x), gamma) * np.log(x)
print(f"Cross Entropy loss = {c1}")
print(f"Focal Loss = {f1}")

print("\nCase 2 ::: Hard Example(Misclassified, y=1, x~0)")
print("Foreground with x = 0.1")
x = 0.1
c2 = -np.log(x)
f2 = -alpha * np.power((1 - x), gamma) * np.log(x)
print(f"Cross Entropy Loss = {c2}")
print(f"Focal Loss = {f2}")

print("\nLoss ratio:")
print(f"Loss ratio in case 1 :: CE/FE = {c1/f1}")
print(f"Loss ratio in case 2 :: CE/FE = {c2/f2}")
```
