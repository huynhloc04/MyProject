```python
# Download CelebA dataset
!mkdir -p /content/datasets
!kaggle datasets download -d jessicali9530/celeba-dataset -p /content/datasets

# Unzip data
from zipfile import ZipFile
from tqdm import tqdm
dataPath = '/content/datasets/celeba-dataset.zip'
extractionPath = '/content/datasets'
with ZipFile(dataPath, 'r') as zf:
   for member in tqdm(zf.infolist(), desc='Extracting'):
     try:
       # Extract all the contents of zip file in the data directory
       zf.extract(member, extractionPath)
     except zipfile.error as e:
       pass
```
